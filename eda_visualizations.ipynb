```python
import warnings
warnings.filterwarnings('ignore')

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('CVD_data.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
plt.figure()

data['Heart_Disease'].value_counts().plot.bar()

plt.title('Distribution of Heart Disease')
plt.xlabel('Heart Disease Class')
plt.ylabel('Counts of Heart Disease')
plt.xticks(rotation = 0)
```

```python
data.head()
```

```python

data['Exercise'] = data['Exercise'].map({'Yes':1,'No':0})

data['Heart_Disease'] = data['Heart_Disease'].map({'Yes':1,'No':0})

data['Skin_Cancer'] = data['Skin_Cancer'].map({'Yes':1,'No':0})

data['Other_Cancer'] = data['Other_Cancer'].map({'Yes':1,'No':0})

data['Depression'] = data['Depression'].map({'Yes':1,'No':0})

data['Arthritis'] = data['Arthritis'].map({'Yes':1,'No':0})

data['Sex'] = data['Sex'].map({'Male':1,'Female':0})

data['Smoking_History'] = data['Smoking_History'].map({'Yes':1,'No':0})	
```

```python
data.head()
```

```python
data['General_Health'].unique()
```

```python
data['General_Health'] = data['General_Health'].map({'Poor':0, 
                                                     'Very Good':1, 
                                                     'Good':2, 
                                                     'Fair':3, 
                                                     'Excellent':4})
```

```python
data['Checkup'].unique()
```

```python
data['Checkup'] = data['Checkup'].map({'Within the past 2 years':1, 
                                       'Within the past year':1,
                                       '5 or more years ago':0, 
                                       'Within the past 5 years':1, 
                                       'Never':0})
```

```python
data['Diabetes'].unique()
```

```python
data['Diabetes'] = data['Diabetes'].map({'No':0, 
                                         'Yes':1, 
                                         'No, pre-diabetes or borderline diabetes':0,
                                         'Yes, but female told only during pregnancy':1})
```

```python
data['Age_Category'].unique()
```

```python
age_mapping = {
    '18-24': (18 + 24) / 2,
    '25-29': (25 + 29) / 2,
    '30-34': (30 + 34) / 2,
    '35-39': (35 + 39) / 2,
    '40-44': (40 + 44) / 2,
    '45-49': (45 + 49) / 2,
    '50-54': (50 + 54) / 2,
    '55-59': (55 + 59) / 2,
    '60-64': (60 + 64) / 2,
    '65-69': (65 + 69) / 2,
    '70-74': (70 + 74) / 2,
    '75-79': (75 + 79) / 2,
    '80+': 85
}
```

```python
data['Age_Category'] = data['Age_Category'].map(age_mapping)
```

```python
data.isnull().sum()
```

```python
data.to_csv('CVD_cleaned.csv', index=False)
```

```python
# Group the data by age group and count the number of heart disease cases
heart_disease_count = data.groupby('Age_Category')['Heart_Disease'].sum()

plt.figure(figsize=(10,8))

# Create a bar chart
plt.bar(heart_disease_count.index, heart_disease_count.values)

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Heart Disease Count')
plt.title('Heart Disease Count Among Age Groups')

# Show the plot
plt.show()
```

```python
# Group the data by heart disease status and count the number of diabetes cases
diabetes_count = data.groupby('Heart_Disease')['Diabetes'].sum()

# Create a bar chart
plt.bar(diabetes_count.index, diabetes_count.values)

# Modify the x-axis to display only 0 or 1
plt.xticks([0, 1], ['No Diabetes', 'Diabetes'])

# Add labels and title
plt.xlabel('Heart Disease Status')
plt.ylabel('Diabetes Count')
plt.title('Diabetes Count Based on Heart Disease Status')

# Show the plot
plt.show()
```

```python
data.corr()
```

```python
# Calculate all the correlations between each independent variables
corr_df = data.corr()
mask = np.triu(corr_df)

plt.figure(figsize=(20,20))
sns.heatmap(corr_df.round(2), annot=True, vmax=1, vmin=-1, cmap='coolwarm', mask=mask, center=0)
plt.show()
```

```python
data.corr()[['Heart_Disease']].sort_values(by='Heart_Disease', ascending=False)
```

```python
plt.figure()

heatmap = sns.heatmap(data.corr()[['Heart_Disease']].sort_values(by='Heart_Disease', ascending=False), vmin=-1, vmax=1, annot=True, cmap='coolwarm')

heatmap.set_title('Features Correlating with Heart Disease', fontdict={'fontsize':18}, pad=16)
```
